```python
# =========================
# 1. Importar librerías
# =========================
import pandas as pd
import re
```

```python
df = pd.read_csv('users.csv', dtype=str)  # dtype=str evita que pandas convierta automáticamente

print("Dataset original:")
display(df.head())
print(f"Filas originales: {len(df)}")
```

```python
# =========================
# 3. Función para detectar notación científica
# =========================
def tiene_notacion_cientifica(valor):
    if pd.isna(valor):
        return False
    # Detecta patrones como 1.23E+10 o 5E-3
    return bool(re.search(r'^\d+(\.\d+)?[eE][+-]?\d+$', str(valor)))
```

```python
 #=========================
# 4. Detectar filas con notación científica
# =========================
mask = df.apply(lambda col: col.map(tiene_notacion_cientifica)).any(axis=1)

print(f"Filas con notación científica: {mask.sum()}")
```

```python
# =========================
# 5. Eliminar esas filas
# =========================
df_clean = df[~mask].copy()

print(f"Filas después de limpiar: {len(df_clean)}")
```

```python
# =========================
# 6. Guardar resultado
# =========================
df_clean.to_csv('archivo_limpio.csv', index=False)

print("Archivo limpio guardado como 'archivo_limpio.csv'")
```

```python
#normalize dates


# =========================
# 2. Cargar dataset
# =========================
df = pd.read_csv("campaigns.csv")

# =========================
# 3. Convertir fechas
# =========================
# dayfirst=True es clave para formato latino (dd/mm/yyyy)
df["start_date"] = pd.to_datetime(df["start_date"], dayfirst=True, errors="coerce")
df["end_date"] = pd.to_datetime(df["end_date"], dayfirst=True, errors="coerce")

# =========================
# 4. Unificar formato (ej: YYYY-MM-DD)
# =========================
df["start_date"] = df["start_date"].dt.strftime("%Y-%m-%d")
df["end_date"] = df["end_date"].dt.strftime("%Y-%m-%d")

# =========================
# 5. Ver resultado
# =========================
print(df)

# =========================
# 6. Guardar limpio
# =========================
df.to_csv("archivo_fechas_limpias.csv", index=False)
```
